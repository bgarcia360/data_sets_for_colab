```python
data = pd.read_csv('athlete_events.csv')
data.head()
```

<h1> Data Clean<h1>

Since this data set has summer and winter, and I am only interested in the summer data sets, I can filter the data and drop any N/A in the age column. 

```python
summer = data[data['Season']== 'Summer']
summer = summer.dropna(axis=0,subset=['Age'])
summer.head()
```

<h1> Age By Olympic History</h1>
<h2>Who is the youngest olympian?</h2>

```python

summer.sort_values(by =['Age']).head(1)
```

<h2> Who is the Oldest olympian? </h2>

```python
summer.sort_values(by=['Age']).tail(1)
```

<h2>Youngest olympian by sport</h2>

```python
by_sport = summer.groupby('Sport')['Age'].min()
print(by_sport.sort_values())
```

<h2> Oldest olympian by sport</h2>

```python
by_sport = summer.groupby('Sport')['Age'].max()
print(by_sport.sort_values(ascending=False))
```

<H2>Boxplot Graph by sport</H2>

```python
by_sport_box = sns.boxplot(x='Sport',y='Age',data=summer)
plt.xticks(rotation=90)
```

<H1>What is the average age from 1896 to 2016?</H1>

```python
Overall_avg_age = summer['Age'].mean()
Overall_avg_age_by_gender = summer.groupby('Sex')['Age'].mean()
print("What is the average age? The average age is "+str(Overall_avg_age)+"\n")
print("What is the average age by gender? The average age by sex is \n" +str(Overall_avg_age_by_gender))

#avg_age_by_country = summer.groupby('NOC')['Age'].mean()
#print(avg_age_by_country.sort_values().head(1))
#print(avg_age_by_country.sort_values().tail(1))
#print("by_sport.sort_values().head(1))
```

<H2> What is the average age by sport?</h2>

```python
by_sport=summer.groupby("Sport")['Age'].mean()
print(by_sport.sort_values())
```

<h2> What is the average age of olympian by year? </h2>

```python
by_year =summer.groupby('Year')['Age'].mean().reset_index(name='avg age')
print(by_year.sort_index())
_age=sns.lineplot(y=by_year["avg age"],x=by_year['Year'])
```
